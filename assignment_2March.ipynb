```python
import logging 
import matplotlib.pyplot as plt

logging.basicConfig(filename="2mar.log",level=logging.INFO)
```

```python
# Ans no1

logging.info("five plots that can be plotted using the Pyplot module of Matplotlib.")

'''Matplotlib is a Python library
 used for data visualization.'''
'''Matplotlib is used extensively in scientific research, data analysis, and engineering applications.'''

'''common plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot: It is a plot of a series of data points connected by straight line segments. 
It is commonly used to visualize trends and patterns in data.

Scatter plot: It is a plot of two variables that shows the relationship between them.
Each data point is represented by a dot, and the position of the dot on the plot 
represents the value of the two variables.

Bar plot: It is a plot of categorical data that shows the frequency or proportion of each category. 
It is commonly used to compare the values of different categories.

Histogram: It is a plot of numerical data that shows the distribution of the data. 
It is commonly used to visualize the frequency distribution of a dataset.

Pie chart: It is a plot of categorical data that shows the proportion of each category
as a slice of a pie. It is commonly used to show the relative sizes of different categories.
'''
```

```python
# Ans no2

'''Scatter plot: It is a plot of two variables that shows the relationship between them.
Each data point is represented by a dot, and the position of the dot on the plot 
represents the value of the two variables.'''

logging.info("creating scatterplot")
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.title("title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(x,y,c='r',alpha=.5,s=300)
```

```python
# Ans no3

'''In Matplotlib, a subplot is a way of arranging multiple plots on a single figure.
Each subplot is a separate plot within the figure, and they can be arranged in a variety of ways.'''


logging.info("creating subplot")
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the four lines on their respective subplots
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a main title to the entire figure
fig.suptitle('Four Line Plots')

# Display the figure
plt.show()
```

```python
# Ans no4

logging.info("creating horizontal,vertical barplot")
'''
Bar plot: It is a plot of categorical data that shows the frequency or proportion of each category. 
It is commonly used to compare the values of different categories.'''

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

logging.info("creating vertical barplot")
plt.bar(company,profit)
# plt.barh(company,profit)
```

```python
logging.info("creating horizontal barplot")
plt.barh(company,profit)
```

```python
# Ans no5

'''In Matplotlib, a boxplot is a graphical representation of the distribution of data based on five number summary:
 minimum, first quartile (Q1), median, third quartile (Q3), and maximum. 
 Boxplots are commonly used to visualize the spread and skewness of a dataset,
and to identify potential outliers.'''

logging.info("a box plot,it's use ,plotting a box plot.")
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the two boxes
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# print(box1)
# Create a list of the data to be plotted
data = [box1, box2]

# Create a figure and axes for the plot
fig, ax = plt.subplots()

# Plot the box plot using the boxplot() function
ax.boxplot(data)

# Add labels to the plot
ax.set_title('Box Plot')
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')

# Display the plot
plt.show()
```
