```python
# Ans no10


data={'Sales':[34,34,53,24,12,34,34,53,24,12],"Date":['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05','2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05']}
df=pd.DataFrame(data)
import pandas as pd

def add_moving_average(df):
    
    ma = df['Sales'].rolling(window=7).mean()
  
    df['MovingAverage'] = ma
    return df

df = add_moving_average(df)
df
```

```python
# Ans no 11

# logging.info("answering th qus no11")
import pandas as pd
data={"Date":['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05']}
df=pd.DataFrame(data)
def create_day(df):
    l=pd.date_range(start='2023-01-01',end='2023-01-05')
    new=pd.Series(l)
    new1= new.dt.day_name()
    df['Weekday']=new1
    

create_day(df)
df.head()
```
