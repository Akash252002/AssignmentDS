#Ans no2

```python
mean_glucose = df['Glucose'].mean()
mean_blood_pressure = df['BloodPressure'].mean()
mean_skin_thickness = df['SkinThickness'].mean()
mean_insulin = df['Insulin'].mean()
mean_bmi = df['BMI'].mean()

# replace NaN values with mean for each column
df['Glucose'] = df['Glucose'].replace(0,mean_glucose)
df['BloodPressure'] = df['BloodPressure'].replace(0,mean_blood_pressure)
df['SkinThickness'] = df['SkinThickness'].replace(0,mean_skin_thickness)
df['Insulin'] = df['Insulin'].replace(0,mean_insulin)
df['BMI'] = df['BMI'].replace(0,mean_bmi)
```

```python

# Identify the quartiles and interquartile range for 'Insulin'
Q1 = df['Insulin'].quantile(0.25)
Q3 = df['Insulin'].quantile(0.75)
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Replace the outliers with the median value of the variable
median = df['Insulin'].median()
df.loc[(df['Insulin'] < lower_bound) | (df['Insulin'] > upper_bound), 'Insulin'] = median

# # Save the cleaned dataset

```

#Ans no3

```python
X=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(  X, y, test_size=0.33, random_state=42)
```

#Ans no4

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
clf = tree.DecisionTreeClassifier()
clf.fit(X_train,y_train)
# tree.plot_tree(clf)
```

#Ans no4

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import log_loss
from sklearn.model_selection import GridSearchCV

# Initialize the decision tree classifier with default hyperparameters
dtc = DecisionTreeClassifier(random_state=42)

# Use cross-validation to optimize the hyperparameters
params = {
    'max_depth': [None, 4, 5, 10],
    'min_samples_split': [10 , 10, 50],
    'min_samples_leaf': [1, 5, 10]
}
grid_search = GridSearchCV(dtc, params, cv=5, scoring='neg_log_loss')
grid_search.fit(X_train, y_train)

# # Print the best hyperparameters and their associated score
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Score: ", -grid_search.best_score_)

```

```python
# Train the model with the best hyperparameters on the entire training set
# best_dtc = grid_search.best_estimator_


best_dtc = DecisionTreeClassifier(max_depth= 4, min_samples_leaf =1, min_samples_split= 10)
best_dtc.fit(X_train, y_train)
# Evaluate the model on the test set
y_pred = best_dtc.predict(X_test)
log_loss_score = log_loss(y_test, y_pred, eps=1e-15)
print("Log Loss Score: ", log_loss_score)
```

#Ans no5

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, confusion_matrix

# Calculate evaluation metrics
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Calculate the false positive rate (fpr) and true positive rate (tpr) for ROC curve
probs = clf.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, probs)
roc_auc = auc(fpr, tpr)

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)
# Print evaluation metrics and confusion matrix
print("Accuracy: {:.4f}".format(acc))
print("Precision: {:.4f}".format(prec))
print("Recall: {:.4f}".format(rec))
print("F1 Score: {:.4f}".format(f1))
print("\nConfusion Matrix:\n", cm)
```

#Ans no6

```python
# Plot ROC curve
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc="lower right")
plt.show()

```
